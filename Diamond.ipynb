```python
X = df1.drop('price', axis = 1)
```

```python
X
```

```python
Y = df1['price']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y) 
```

```python
X_train.to_pickle("xtrain", compression="xz") 
```

```python
y_train.to_pickle("ytrain", compression="xz")
```

```python
X_test.to_pickle("xtest", compression = "xz")
```

```python
y_test.to_pickle("ytest", compression = "xz")
```

```python
X_test.shape 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
explained_variance_score(y_test, y_pred)
```

```python
sgd = SGDRegressor(max_iter=10_000, n_iter_no_change=100)
```

```python
sgd.fit(X_train, y_train)
sgd.score(X_test, y_test)
```
